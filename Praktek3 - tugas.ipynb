```python
import pandas as pd

mark = pd.read_csv('BahanPraktikum1.csv')

mark['Kategori Tinggi'] = mark['Tinggi Badan'].apply(lambda x: 'Tinggi' if x > 160 else 'Pendek')

print(mark)
```

```python
import numpy as np

mark['Jurusan'] = np.full(len(mark), 'Infor24')
mark['Fakultas'] = np.full(len(mark), 'FTI')

print(mark)
```

```python
mark = mark.drop(columns=['Fakultas'])
print(mark)
```

```python
namadangender = mark.iloc[:, 0:2]
print(namadangender)
```

```python
angkatantb = mark.iloc[:, 2:4]
print(angkatantb)
```

```python
# Menggabungkan dua DataFrame 
gabungan = pd.concat([namadangender, angkatantb], axis=1)

# Menampilkan DataFrame gabungan
print(gabungan)
```
